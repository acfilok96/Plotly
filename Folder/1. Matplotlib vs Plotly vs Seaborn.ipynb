```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as pl

import numpy as np
```

```python
x = np.array([i for i in range(10)]) # np.random.randint(1,10,size=(10,))
y = np.array([i**2 for i in x])
print(x)
print()
print(y)
```

```python
plt.plot(x,y,color='red')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
fig = sns.lineplot(x, y, color='red')
fig.set(xlabel="X-axis", ylabel="Y-axis")
fig.grid()
fig.set_title("Line Plot")
```

```python
fig = sns.lineplot(data = [x,y],  color='red')
fig.set(xlabel="X-axis", ylabel="Y-axis")
fig.grid()
fig.set_title("Line Plot")
```

```python
import plotly.express as px
fig = px.line(x=x, y=y, title="Line Plot")
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(go.Scatter(x=x,y=y))
fig.show()
```

```python
fig = go.Figure(go.Scatter(x=x,y=y))
fig.update_layout(title='Line Plot', xaxis_title="X-axis", yaxis_title="Y-axis")
fig.show()
```

```python
x1,y1 = [1,2,3,4,5],[5,6,7,4,2]
x2,y2 = [1,2,3,4,5],[6,3,4,1,2]
x3,y3 = [1,2,3,4,5],[6,2,6,8,3]

fig = go.Figure()

fig.add_trace(go.Scatter(x=x1,y=y1))
fig.add_trace(go.Scatter(x=x2,y=y2))
fig.add_trace(go.Scatter(x=x3,y=y3))

fig.update_layout(title="Line Plot",xaxis_title="x-axis",yaxis_title="y-axis")

fig.show()
```

```python
x1,y1 = [1,2,3,4,5],[5,6,7,4,2]
x2,y2 = [1,2,3,4,5],[6,3,4,1,2]
x3,y3 = [1,2,3,4,5],[6,2,6,8,3]

fig = go.Figure()

fig.add_trace(go.Scatter(x=x1,y=y1,name='line1',mode='lines'))
fig.add_trace(go.Scatter(x=x2,y=y2,name='line2',mode='markers'))
fig.add_trace(go.Scatter(x=x3,y=y3,name='line3',mode='lines+markers'))

fig.update_layout(title="Line Plot",xaxis_title="x-axis",yaxis_title="y-axis")

fig.show()
```

```python
x = list(range(1,10))
y = [1,2,None,5,8,2,None,None,9,4]
z = [5,7,4,None,1,9,0,3,1,None]
fig =go.Figure()
fig.add_trace(go.Scatter(x=x,y=y,name="<b>No</b> Gaps",connectgaps=True))
fig.add_trace(go.Scatter(x=x,y=z,name="Gaps"))
fig.update_layout(title="Line Plot",xaxis_title="X-axis",yaxis_title="Y-axis")
fig.show()
```
